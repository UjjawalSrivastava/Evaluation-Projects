#### Outlier Detection

```python
# Checking outliers.
plt.figure(figsize=(10, 6), facecolor='White')
plotnumber = 1
for col in df:
    if plotnumber <=16:
        ax = plt.subplot(4,4,plotnumber)
        sns.boxplot(df[col], palette='Set2_r')
        plt.xlabel(col, fontsize= 15)
        plt.yticks(rotation=0, fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
# Detecting and removing outliers using the IQR method.

# Function to remove outliers
def removing_outliers_from_iqr_method(df, columns):
    new_df = df.copy()
    for col in columns:
        Q1 = new_df[col].quantile(0.25)
        Q3 = new_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        new_df = new_df[(new_df[col] >= lower_bound) & (new_df[col] <= upper_bound)]
    return new_df

print("\nData shape before and after outlier removal:")
print("Before:", df.shape)

# Apply outlier removal
df = removing_outliers_from_iqr_method(df, df.columns)

print("After:", df.shape)
```

```python
# Checking the outliers after removing all the possible outliers.
plt.figure(figsize=(10, 6), facecolor='White')
plotnumber = 1
for col in df:
    if plotnumber <=16:
        ax = plt.subplot(4,4,plotnumber)
        sns.boxplot(df[col], palette='Set2_r')
        plt.xlabel(col, fontsize= 15)
        plt.yticks(rotation=0, fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
# Separating feature and label
x = df.drop('Price range', axis=1)
y = df['Price range']
```

```python
# Feature Scaling using Standard Scalarization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
```

```python
# Checking Variance Inflation Factor(VIF)
# Checking VIF values to confirm whether multicollinearity exists or not.
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['VIF Values'] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]
vif['Feature Values'] = x.columns
```

#### So, we can clearly see that there is no multicollinearity present, so we do not need to drop columns.

```python
# Classification Algorithmfrom
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Feature selection for regression (exclude the target 'Price range')
x_reg = x
y_reg = y
```

```python
# Finding the best R-squared at random state after that you can comment these line of code.
max_r2 = 0
best_random_state = 0

# Loop through random states to find the best one
for i in range(1, 200):
    x_train, x_test, y_train, y_test = train_test_split(x_reg, y_reg, test_size=0.30, random_state=i)
    RFR_regressor = RandomForestRegressor()
    RFR_regressor.fit(x_train, y_train)
    
    # Predict and evaluate the model
    y_pred_reg = RFR_regressor.predict(x_test)
    r2 = r2_score(y_test, y_pred_reg)
    
    # Update the best R-squared and random state
    if r2 > max_r2:
        max_r2 = r2
        best_random_state = i

# Print the best R-squared and corresponding random state
print("Best R-squared:", max_r2, "at random_state:", best_random_state)
```

```python
# Split data into train and test sets.
x_reg_train, x_reg_test, y_reg_train, y_reg_test = train_test_split(x_reg, y_reg, test_size=0.3, random_state=best_random_state)

# Train a Random Forest Regressor
RFR_regressor = RandomForestRegressor()
RFR_regressor.fit(x_reg_train, y_reg_train)
```

```python
# Predictions and Evaluation
y_reg_pred = RFR_regressor.predict(x_reg_test)
print('RMSE:', mean_squared_error(y_reg_test, y_reg_pred, squared=False))
print('R-squared:', r2_score(y_reg_test, y_reg_pred))
```

```python
import joblib

# Saving the model.
joblib.dump(RFR_regressor, 'zomato_restaurant.pkl')
```

# Separating the target two column for model.

```python
x_reg_target_two = df.drop('Average Cost for two', axis=1)
y_reg_target_two = df['Average Cost for two']
```

```python
# Finding the best R-squared at random state after that you can comment these line of code.
max_r2 = 0
best_random_state = 0

# Loop through random states to find the best one
for i in range(1, 200):
    x_train, x_test, y_train, y_test = train_test_split(x_reg_target_two, y_reg_target_two, test_size=0.30, random_state=i)
    RFR_regressor = RandomForestRegressor()
    RFR_regressor.fit(x_train, y_train)
    
    # Predict and evaluate the model
    y_pred_reg = RFR_regressor.predict(x_test)
    r2 = r2_score(y_test, y_pred_reg)
    
    # Update the best R-squared and random state
    if r2 > max_r2:
        max_r2 = r2
        best_random_state = i

# Print the best R-squared and corresponding random state
print("Best R-squared:", max_r2, "at random_state:", best_random_state)
```

```python
# Split data into train and test sets.
x_train_two, x_test_two, y_train_two, y_test_two = train_test_split(x_reg_target_two, y_reg_target_two, test_size=0.3, random_state=46)

# Train a Random Forest Regressor
RFR_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
RFR_regressor.fit(x_train_two, y_train_two)
```

```python
# Predictions and Evaluation
y_reg_pred = RFR_regressor.predict(x_test_two)
print('RMSE:', mean_squared_error(y_test_two, y_reg_pred, squared=False))
print('R-squared:', r2_score(y_test_two, y_reg_pred))
```

# Hyperparameter Tuning

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
}

grid_search = GridSearchCV(estimator=RFR_regressor, param_grid=param_grid, cv=5)
grid_search.fit(x_train_two, y_train_two)

# Best parameters
print(f'Best parameters: {grid_search.best_params_}')
```

```python
# Use the best model for predictions
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(x_test_two)
```

```python
# Evaluate the tuned model
mse_best = mean_squared_error(y_test_two, y_pred_best)
print(f'Optimized Mean Squared Error: {mse_best}')
print('R-squared:', r2_score(y_test_two, y_pred_best))
```
